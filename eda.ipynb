```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Ajusta la ruta según tu repo
df = pd.read_csv("./data/raw/dataset_practica_final.csv")

print("Shape:", df.shape)
display(df.head())
```

```python
df.info()
```

```python
df.describe(include="all").T
```

```python
target_dist = df["is_canceled"].value_counts(normalize=True)

display(target_dist)

sns.countplot(x="is_canceled", data=df)
plt.title("Distribución de la variable objetivo (is_canceled)")
plt.show()
```

```python
missing = (
    df.isnull()
      .mean()
      .sort_values(ascending=False)
      .to_frame("missing_ratio")
)

missing[missing["missing_ratio"] > 0]
```

```python
missing = missing[missing["missing_ratio"] > 0]
sns.barplot(
    x=missing.index,
    y="missing_ratio",
    data=missing[missing["missing_ratio"] > 0]
)
plt.xticks(rotation=90)
plt.title("Proporción de valores nulos por variable")
plt.show()
```

```python
print("Filas duplicadas:", df.duplicated().sum())
```

```python
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(bins=30, figsize=(18, 14))
plt.suptitle("Distribución de variables numéricas", y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(x=df["lead_time"], ax=axes[0])
axes[0].set_title("Boxplot - lead_time")

sns.boxplot(x=df["adr"], ax=axes[1])
axes[1].set_title("Boxplot - adr")

plt.show()
```

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    print(f"\n{col}")
    display(df[col].value_counts().head(10))
```

```python
sns.countplot(x="hotel", hue="is_canceled", data=df)
plt.title("Cancelaciones por tipo de hotel")
plt.show()
```

```python
sns.boxplot(x="is_canceled", y="lead_time", data=df)
plt.title("Lead time vs cancelación")
plt.show()
```

```python
sns.countplot(x="is_repeated_guest", hue="is_canceled", data=df)
plt.title("Cancelaciones según cliente recurrente")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(
    y="market_segment",
    hue="is_canceled",
    data=df,
    order=df["market_segment"].value_counts().index
)
plt.title("Cancelaciones por segmento de mercado")
plt.show()
```

```python
corr = df[num_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matriz de correlación (variables numéricas)")
plt.show()
```

```python
df["arrival_date_month"] = pd.Categorical(
    df["arrival_date_month"],
    categories=[
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ],
    ordered=True
)

cancel_by_month = df.groupby("arrival_date_month")["is_canceled"].mean()

cancel_by_month.plot(marker="o")
plt.title("Tasa de cancelación por mes")
plt.ylabel("Cancel rate")
plt.show()
```

### Principales conclusiones del EDA

- El dataset está **moderadamente desbalanceado**, por lo que métricas como F1-score o AUC-ROC son más adecuadas que accuracy.
- `lead_time` es una de las variables con mayor poder explicativo: reservas con mayor antelación se cancelan más.
- Clientes repetidos presentan una **menor tasa de cancelación**.
- Existen **outliers relevantes en `adr` y `lead_time`**, que deben tratarse antes del modelado.
- Variables categóricas como `market_segment`, `deposit_type` y `customer_type` parecen muy informativas.
- Se requerirá:
  - Imputación de nulos
  - Encoding de categóricas
  - Escalado de numéricas
